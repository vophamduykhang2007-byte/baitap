```python
# 1. Cài đặt thư viện
!pip install gradio pandas scikit-learn -q

import gradio as gr
import pandas as pd
from sklearn.linear_model import Perceptron
from sklearn.preprocessing import StandardScaler

# 2. Huấn luyện mô hình
def train_perceptron_model():
    try:
        df = pd.read_csv('/content/diabetes.csv')
        X = df.drop('Outcome', axis=1)
        y = df['Outcome']
        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X)
        model = Perceptron(max_iter=1000, tol=1e-3, random_state=42)
        model.fit(X_scaled, y)
        return model, scaler, X.describe()
    except Exception as e:
        print(f"Lỗi: Hãy nhớ upload file diabetes.csv! ({e})")
        return None, None, None

model, scaler, stats = train_perceptron_model()

# 3. Hàm dự đoán
def predict_diabetes(preg, glu, bp, skin, ins, bmi, dpf, age):
    input_data = pd.DataFrame([[preg, glu, bp, skin, ins, bmi, dpf, age]],
                               columns=['Pregnancies', 'Glucose', 'BloodPressure',
                                        'SkinThickness', 'Insulin', 'BMI',
                                        'DiabetesPedigreeFunction', 'Age'])
    input_scaled = scaler.transform(input_data)
    prediction = model.predict(input_scaled)[0]
    return "🚨 CÓ NGUY CƠ" if prediction == 1 else "✅ AN TOÀN"

# 4. Giao diện (Đã fix DeprecationWarning)
with gr.Blocks() as demo:
    gr.Markdown(f"# 🩺 Chẩn đoán Tiểu đường - Perceptron\n**SV: Võ Phạm Duy Khang**")
    with gr.Row():
        with gr.Column():
            p = gr.Number(label="Pregnancies", value=3)
            g = gr.Slider(0, 200, label="Glucose", value=120)
            b = gr.Slider(0, 140, label="BloodPressure", value=70)
            s = gr.Slider(0, 100, label="SkinThickness", value=20)
            i = gr.Slider(0, 900, label="Insulin", value=80)
            bmi = gr.Number(label="BMI", value=25.0)
            d = gr.Number(label="DPF", value=0.5)
            a = gr.Slider(1, 100, label="Age", value=30)
            btn = gr.Button("DỰ ĐOÁN", variant="primary")
        with gr.Column():
            output = gr.Textbox(label="Kết quả")

    btn.click(fn=predict_diabetes, inputs=[p, g, b, s, i, bmi, d, a], outputs=output)

# CHÚ Ý: Link sẽ hiện ở đây sau khi chạy
demo.launch(share=True, debug=True)
```
